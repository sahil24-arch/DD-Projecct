```python
import pandas as pd
```

```python
#importing file containig Abstract text
abs=pd.read_csv('Abstracts.csv')
pd.set_option('display.max_colwidth', None)
abs.head()
```

```python
#Importing file containig drug and disease information
drug=pd.read_csv('Drugs.csv')
drug.head()
```

```python
#Converting drug dataframe to csv
drug_dict=dict(zip(drug.term, drug.type))
```

```python
df=abs.copy()
```

```python
# Function to extract disease name from the text
def extract_disease(s):
    l=[]
    for i in drug_dict.keys():

        if s.find(i)==-1:
            continue
        else:
            if drug_dict[i]=='Disease':
                l.append(i)
            else:
                return ''
    return ";;".join(l)
```

```python
# Function to extract drug name from the text
def extract_drug(s):
    l=[]
    for i in drug_dict.keys():
        
        if s.find(i)==-1:
            continue
        else:
            if drug_dict[i]=='Drug':
                l.append(i)
            else:
                return ''
    return ";;".join(l)
```

```python
#Extracting disease and drug name and filling it in respective columns
df['disease']=df['abstract'].apply(extract_disease)
df['drug']=df['abstract'].apply(extract_drug)
```

```python
df.head()
```

```python
#From the above output,we can see that some words are being repeated which are unnecessary
#For this we need to perform some sort of cleaning to remove unnecesary words
```

```python
#Function to avoid repeatation
def clean(s):
    l=s.split(';;')
    l1=l.copy()
    for i in l:
        for j in l1:
            try:
                if i == j:continue
                if i in j:
                    l1.remove(i)
                  
            except Exception as e: return ";;".join(l1) 
              
    return ";;".join(l1) 
```

```python
df1=df.copy()
```

```python
#Cleaning of disease and drug column
df1['disease']=df1['disease'].apply(clean)
df1['drug']=df1['drug'].apply(clean)
```
